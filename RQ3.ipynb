```python
import os
import json
import numpy as np
import matplotlib.pyplot as plt

# Folder path containing JSON files
folder_path = '/Users/ayahmiqdad/Desktop/Thesis literutre 2024/icpe-data-challenge-jmh/timeseries'

# Function to extract performance measurements from the first fork of a JSON file
def extract_first_fork_measurements(json_filepath):
    with open(json_filepath, 'r') as json_file:
        data = json.load(json_file)
        first_fork_measurements = data[0]  # Assuming measurements from the first fork are at index 0
        return first_fork_measurements

# Function to calculate statistical measures
def calculate_statistics(data):
    # Exclude the first measurement (steady state)
    data = data[6:]
    mean_value = np.mean(data)
    variance_value = np.var(data, ddof=1)  # Use Bessel's correction for variance
    std_deviation_value = np.std(data)
    range_value = np.max(data) - np.min(data)
    return mean_value, variance_value, std_deviation_value, range_value


# Plotting
def plot_measurements(json_filepath, measurements, mean, std_deviation):
    
    # Normalize the data
    min_value = np.min(measurements[1:])
    max_value = np.max(measurements[1:])
    normalized_measurements = (measurements[1:] - min_value) / (max_value - min_value)
    
    plt.figure(figsize=(14, 12))
    plt.plot(normalized_measurements, label='Normalized Measurements', color='blue')
    plt.axhline(y=np.mean(normalized_measurements), color='red', linestyle='--', label='Mean')
    std_deviation_1 = np.std(normalized_measurements)
    plt.fill_between(range(len(normalized_measurements)), 
                     np.mean(normalized_measurements) - std_deviation_1, 
                     np.mean(normalized_measurements) + std_deviation_1,
                     color='orange', alpha=0.2, label='1 Standard Deviation')
    std_deviation_3 = 3 * np.std(normalized_measurements)
    plt.fill_between(range(len(normalized_measurements)), 
                     np.mean(normalized_measurements) - std_deviation_3, 
                     np.mean(normalized_measurements) + std_deviation_3,
                     color='green', alpha=0.2, label='3 Standard Deviations')
    
    # Increase font sizes for labels and title
    plt.xlabel('Time', fontsize=18)
    plt.ylabel('Measurement (Normalized)', fontsize=18)
    plt.title('Normalized Performance Measurements: {}'.format(json_filepath), fontsize=20)
    
    # Increase font size for the legend
    plt.legend(fontsize=16)
    
    # Increase font size for the tick labels
    plt.xticks(fontsize=14)
    plt.yticks(fontsize=14)
    
    # Display grid
    plt.grid(True)
    
    # Show plot
    plt.show()
# Loop through all JSON files in the folder
for filename in os.listdir(folder_path):
    if filename.endswith('.json'):
        json_filepath = os.path.join(folder_path, filename)
        # Extract performance measurements from the JSON file
        first_fork_measurements = extract_first_fork_measurements(json_filepath)
        # Calculate statistical measures
        mean, variance, std_deviation, range_val = calculate_statistics(first_fork_measurements)
        # Print statistical measures
        print("File:", filename)
        print("Mean from steady state to end of data:", mean)
        print("Variance from steady state to end of data (with Bessel's correction):", variance)
        print("Standard Deviation from steady state to end of data:", std_deviation)
        print("Range from steady state to end of data:", range_val)
        # Plot measurements
        plot_measurements(json_filepath, first_fork_measurements, mean, std_deviation)
        
# Save the DataFrame to a CSV file
output_csv_filepath = '/Users/ayahmiqdad/Desktop/Thesis literutre 2024/icpe-data-challenge-jmh/statistics.csv'
df.to_csv(output_csv_filepath, index=False)


```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read the CSV file
csv_filepath = '/Users/ayahmiqdad/Desktop/Thesis literutre 2024/icpe-data-challenge-jmh/statistics1.csv'
data = pd.read_csv(csv_filepath)

# Extract variance and standard deviation based on the columns named 'Variance' and 'Standard Deviation'
variance = data['Variance']
std_deviation = data['Standard Deviation']

# Plotting
plt.figure(figsize=(14, 8))

# Plot density for variance
sns.kdeplot(variance, color='blue', fill=True, label='Variance Density')

# Plot density for standard deviation
sns.kdeplot(std_deviation, color='red', fill=True, label='Standard Deviation Density', alpha=0.5)

plt.xlabel('Value', fontsize=14)
plt.ylabel('Density', fontsize=14)
plt.title('Density Plot: Variance and Standard Deviation', fontsize=16)
plt.legend()
plt.grid(True)

plt.show()

# Create box plot for variance and standard deviation
plt.figure(figsize=(14, 8))
sns.boxplot(data=[variance, std_deviation], palette=['blue', 'red'], notch=True)
plt.xticks([0, 1], ['Variance', 'Standard Deviation'], fontsize=14)
plt.ylabel('Value', fontsize=14)
plt.title('Box Plot: Variance and Standard Deviation', fontsize=16)
plt.grid(True)
plt.show()

# Create histogram for variance and standard deviation
plt.figure(figsize=(14, 8))

# Histogram for variance
plt.hist(variance, bins=30, color='blue', alpha=0.6, label='Variance')

# Histogram for standard deviation
plt.hist(std_deviation, bins=30, color='red', alpha=0.5, label='Standard Deviation')

plt.xlabel('Value', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.title('Histogram: Variance and Standard Deviation', fontsize=16)
plt.legend()
plt.grid(True)

plt.show()
```
